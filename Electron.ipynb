```python
from numpy import *
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
from PlotFuncs import *

plt.style.use('sty.mplstyle')

fig,ax = plt.subplots()

fpath = 'data/limits/Electron/HeavyMediator/'
facecolor=[0.9,0.9,0.9]

PlotBound(ax,fpath+'DarkSide-50.txt',label=r'{\bf DarkSide-50}',label_pos=[23,2e-34],edgecolor='seagreen',facecolor=col_alpha('seagreen',0.5),textcolor='seagreen',rotation=6,zorder=-2)
PlotBound(ax,fpath+'CDMS-HVeV.txt',label=r'{\bf CDMS-HVeV}',label_pos=[40,0.7e-32],edgecolor='gold',facecolor=col_alpha('#993186',0.5),textcolor='gold',rotation=8)
PlotBound(ax,fpath+'SENSEI.txt',label=r'{\bf SENSEI}',label_pos=[0.7,0.07e-32],edgecolor='#993186',facecolor=col_alpha('#993186',0.5),textcolor='#993186',rotation=-30,zorder=-1,fontsize=22)
PlotBound(ax,fpath+'DAMIC-M.txt',label=r'{\bf DAMIC-M}',label_pos=[1.6,0.05e-34],edgecolor='#6e0b3c',facecolor=col_alpha('#6e0b3c',0.5),textcolor='#6e0b3c',rotation=-30,zorder=-2,fontsize=22)
PlotBound(ax,fpath+'EDELWEISS.txt',label=r'{\bf EDELWEISS}',label_pos=[32,1.2e-33],edgecolor='orange',facecolor=facecolor,facealpha=0,textcolor='orange',rotation=7,zorder=1,fontsize=22)
PlotBound(ax,fpath+'PandaX.txt',label=r'{\bf PandaX}',label_pos=[12,0.4e-34],edgecolor='navy',facealpha=0,textcolor='navy',rotation=-23,zorder=1)
PlotBound(ax,fpath+'XENON1T.txt',label=r'{\bf XENON1T}',label_pos=[5,0.1e-34],edgecolor='teal',facecolor=col_alpha('teal',0.5),textcolor='teal',lw=3,rotation=-30,zorder=-10)
PlotBound(ax,fpath+'XENON1T_SolarReflection.txt',label=r'\begin{center} {\bf XENON1T} \linebreak {\bf Solar reflection}\end{center}',label_pos=[5,0.1e-34],facecolor='teal',edgecolor='teal',linestyle='--',facealpha=0.1,textcolor='teal',rotation=-30,zorder=-10,fontsize=14)

PlotBound(ax,fpath+'Projections/OSCURA.txt',label=r'{\bf OSCURA}',label_pos=[50,0.3e-41],edgecolor='gray',facealpha=0,linestyle='--',textcolor='gray',lw=3,text_path_effects=line_background(1,'k'),rotation=8,zorder=1,fontsize=22)


dat1 = loadtxt(fpath+'/Model_ELDER.txt')
dat2 = loadtxt(fpath+'Model_ScalarFreezeIn.txt')
DAT = vstack((flipud(dat1),dat2))
plt.fill_between(DAT[:,0],DAT[:,1],color='gray',hatch='//',alpha=0.1,zorder=-100)
plt.plot(DAT[:,0],DAT[:,1],'-',color='gray',alpha=0.5,zorder=-100)
plt.text(14.5,0.2e-40,r'{\bf Theory target}',color='gray',rotation=-30,path_effects=line_background(1,'k'),alpha=0.5)
plt.text(12.5,0.6e-40,'ELDER',rotation=-31,color='gray',alpha=0.5,fontsize=15)
plt.text(7,9e-39,'Scalar freeze-out',rotation=-22,color='gray',alpha=0.5,fontsize=15)

ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlim([0.5,1e3])
ax.set_ylim([1e-45,1e-28])
ax.set_xlabel('Dark matter mass [MeV]')
ax.set_ylabel(r'DM-electron cross section [cm$^{2}$]',labelpad=10)

locmaj = matplotlib.ticker.LogLocator(base=10.0, subs=(1.0, ), numticks=50)
locmin = matplotlib.ticker.LogLocator(base=10.0, subs=arange(2, 10)*.1,numticks=100)
ax.yaxis.set_major_locator(locmaj)
ax.yaxis.set_minor_locator(locmin)



plt.gcf().text(0.78, 0.82, r'Heavy mediator',ha='center',color='black',multialignment='center',
            bbox=dict(facecolor='w', linewidth=1.5,
                        edgecolor='black', boxstyle='round,pad=0.2'))


MySaveFig(fig,'Electron_HeavyMediator')
```
