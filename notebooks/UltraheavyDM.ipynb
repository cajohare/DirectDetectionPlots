```python
from numpy import *
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
from PlotFuncs import *


def UpperAxis_Msun(ax,tickdir='in',xtick_rotation=0,labelsize=19,xlabel=None,lfs=40,tick_pad=-40,tfs=25,xlabel_pad=10,label_shift=[1,1]):
    m_min,m_max = ax.get_xlim()
    GeV_2_g = 1/5.62e23 # convert GeV to grams
    g_2_Msun = 1/2e33
    ax2 = ax.twiny()
    ax2.set_xscale('log')
    ax2.set_xlabel(xlabel,fontsize=lfs,labelpad=xlabel_pad)
    ax2.tick_params(which='major',direction=tickdir,width=2.5,length=13,pad=tick_pad,rotation=xtick_rotation,labelsize=labelsize)
    ax2.tick_params(which='minor',direction=tickdir,width=1,length=10)

    #ax2.xaxis.set_major_locator(matplotlib.ticker.LogLocator(base=1000,subs=(1.0,),numticks=100))
    #ax2.xaxis.set_minor_locator(matplotlib.ticker.LogLocator(base=10,subs=(1.0,),numticks=100))

    ax2.set_xticks(10.0**arange(-18,-6,3))
    #ax2.set_xticklabels(['ag','fg','pg','ng',r'\textmu g','mg','g','kg','Mg','Gg','Tg','Pg']);
    ax2.set_xticks(10.0**arange(-18,-6-2,1), minor=True)
    ax2.xaxis.set_minor_formatter(matplotlib.ticker.NullFormatter())
    ax2.set_xlim([m_min*GeV_2_g*g_2_Msun,m_max*GeV_2_g*g_2_Msun])
    ax2.text(1e-18/100*label_shift[0],ax2.get_ylim()[1]/200*label_shift[1],r'$M_\odot$',ha='right',fontsize=labelsize)
    plt.sca(ax)
    return



plt.style.use('sty.mplstyle')

fig, ax = plt.subplots()
ax.tick_params(labelsize=24)

fpath = '../data/limits/Ultraheavy/Contact/'

# Plot limits
PlotBound(ax,fpath+'DEAP-3600.txt',y2=nan,edgecolor='#4ff09d',textcolor='#4ff09d',facecolor=col_alpha('#4ff09d',0.6),label=r'{\bf DEAP}',label_pos=[4e18,0.5e-20],fontsize=20)
PlotBound(ax,fpath+'XENON1T.txt',y2=nan,edgecolor='teal',textcolor='teal',facecolor=col_alpha('teal',0.5),label=r'{\bf XENON1T}',label_pos=[3e11,1e-28],rotation=44,fontsize=18)
PlotBound(ax,fpath+'LZ.txt',y2=nan,edgecolor='darkgreen',textcolor='darkgreen',facecolor=[0.0, 0.62, 0.38],label=r'{\bf LZ}',label_pos=[3e16,2e-25],fontsize=20)
PlotBound(ax,fpath+'XENON1T_SS.txt',y2=nan,edgecolor='teal',textcolor='teal',facecolor=col_alpha('teal',0.5))
PlotBound(ax,'../data/limits/Nucleon/SI/LZ_2024.txt',y2=1e-24,edgecolor='darkgreen',textcolor='darkgreen',facecolor=[0.0, 0.62, 0.38],zorder=-100,scale_y=131**4)
PlotBound(ax,'../data/limits/Nucleon/SI/XENON1T.txt',y2=1e-24,edgecolor='teal',textcolor='teal',facecolor=col_alpha('teal',0.5),zorder=-100,scale_y=0.4*131**4)


PlotBound(ax,fpath+'CDMS.txt',y2=nan,edgecolor='y',textcolor='y',facecolor=col_alpha('y',0.5),label=r'{\bf CDMS}',label_pos=[22,2e-30],fontsize=17)
PlotBound(ax,fpath+'CRESST.txt',y2=nan,edgecolor=[0.8, 0.25, 0.33],textcolor=[0.8, 0.25, 0.33],facecolor=col_alpha([0.8, 0.25, 0.33],0.5),label=r'{\bf CRESST}',label_pos=[1e0,0.3e-27],rotation=45,fontsize=15)
PlotBound(ax,fpath+'XQC.txt',y2=nan,edgecolor='orange',textcolor='orange',facecolor=col_alpha('orange',0.5),label=r'{\bf XQC}',label_pos=[5,1e-20],fontsize=20)


PlotBound(ax,fpath+'DAMA.txt',y2=nan,edgecolor='darkslategray',textcolor='darkslategray',facecolor='forestgreen',label=r'{\bf DAMA}',label_pos=[3e5,1e-27],fontsize=20,rotation=23)
PlotBound(ax,fpath+'MAJORANA.txt',y2=nan,edgecolor='#436348',textcolor='#436348',facecolor=col_alpha('#436348',0.8),label=r'{\bf MAJORANA}',label_pos=[2e9,2.5e-24],fontsize=12,rotation=13)

PlotBound(ax,fpath+'Chicago.txt',y2=nan,edgecolor='#873b82',textcolor='#873b82',facecolor=col_alpha('#873b82',0.5),label=r'{\bf Chicago}',label_pos=[2e6,3e-23],fontsize=20,rotation=6)
PlotBound(ax,fpath+'Skylab.txt',y2=nan,edgecolor='#85455e',textcolor='#85455e',facecolor=col_alpha('#85455e',0.5),label=r'{\bf Skylab}',label_pos=[5e10,1e-15],fontsize=25,rotation=47)
PlotBound(ax,fpath+'Ohya.txt',y2=nan,edgecolor='#ab4666',textcolor='#ab4666',facecolor=col_alpha('#ab4666',0.5),label=r'{\bf Ohya}',label_pos=[9e12,0.5e-16],fontsize=25,rotation=48)
PlotBound(ax,fpath+'Mica.txt',y2=nan,edgecolor='brown',textcolor='brown',facecolor=col_alpha('brown',0.6),label=r'{\bf Ancient mica}',label_pos=[1e15,1e-17],fontsize=25)
PlotBound(ax,fpath+'HumanDeath.txt',y2=nan,edgecolor='crimson',textcolor='crimson',facecolor=col_alpha('crimson',0.6),label=r'\begin{center}{\bf Human \linebreak death}\end{center}',label_pos=[2e23,0.25e-5],fontsize=16)
PlotBound(ax,fpath+'WhiteDwarfs.txt',y2=nan,edgecolor='pink',textcolor='pink',facecolor=col_alpha('pink',0.6),label=r'{\bf White dwarfs}',label_pos=[4e36,0.3e-6],fontsize=20)
PlotBound(ax,fpath+'NeutronStars.txt',y2=nan,edgecolor='#f54266',textcolor='#f54266',facecolor=col_alpha('#f54266',0.6),label=r'\begin{center}{\bf Neutron \linebreak stars}\end{center}',label_pos=[1e30,1e-4],fontsize=16,rotation=50)
PlotBound(ax,fpath+'Radar.txt',y2=nan,edgecolor='#8c142c',textcolor='#8c142c',facecolor=col_alpha('#8c142c',0.6),label=r'{\bf Meteors}',label_pos=[3e17,2e-3],fontsize=20,rotation=48)
PlotBound(ax,fpath+'Fireballs.txt',y2=nan,edgecolor='#bd0202',textcolor='#bd0202',facecolor=col_alpha('#bd0202',0.6),label=r'{\bf Fireballs}',label_pos=[0.5e26,3e-3],fontsize=20,rotation=38)

# Finishing touches
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlim([1e-2,1e50])
ax.set_ylim([1e-30,1e10])
ax.set_xlabel('Dark matter mass [GeV/$c^2$]')
ax.set_ylabel(r'SI DM-nucleon cross section [cm$^{2}$]')

ax.yaxis.set_major_locator(matplotlib.ticker.LogLocator(base=1000,subs=(1.0,),numticks=100))
ax.yaxis.set_minor_locator(matplotlib.ticker.LogLocator(base=10,subs=(1.0,),numticks=100))
ax.yaxis.set_minor_formatter(matplotlib.ticker.NullFormatter())
ax.xaxis.set_major_locator(matplotlib.ticker.LogLocator(base=1000,subs=(1.0,),numticks=100))
ax.xaxis.set_minor_locator(matplotlib.ticker.LogLocator(base=10,subs=(1.0,),numticks=100))
ax.xaxis.set_minor_formatter(matplotlib.ticker.NullFormatter())
# ax.axvline(1.22e19,color='gray',linestyle='-',zorder=-100,lw=2)
# ax.text(1.22e19*1.05,1e-28,r'$M_{\rm Pl}$',fontsize=20,ha='left',va='center',color='gray')
UpperAxis_grams(ax)
UpperAxis_Msun(ax)


PlotBound(ax,fpath+'MWSatellites.txt',lw=0,edgecolor='dimgray',textcolor='dimgray',facecolor=col_alpha('dimgray',0.2),label=r'{\bf Gas Cloud}',label_pos=[22,5e-40],path_effects=None,zorder=-100)
PlotBound(ax,fpath+'CMB.txt',lw=0,edgecolor='gray',textcolor='gray',facecolor=col_alpha('gray',0.4),label=r'{\bf CMB}',label_pos=[22,5e-40],path_effects=None,zorder=-99)
PlotBound(ax,fpath+'Microlensing.txt',y2=1e-100,lw=0,edgecolor='gray',textcolor='gray',facecolor=col_alpha('gray',0.4),label=r'{\bf CMB}',label_pos=[22,5e-40],path_effects=None,zorder=-99)

#plt.gcf().text(0.88,0.14,r'{\bf Contact interaction}',fontsize=25,ha='right')

# Black holes
m = array([1e2,1e50])
GeV_2_g = 1/5.62e23 
sigma_BH = pi*(100*2*6.67e-11*m*GeV_2_g*1e-3/3e8**2)**2
plt.fill_between(m,sigma_BH,y2=1e-100,facecolor='dimgray',hatch='\\',edgecolor='k')

# Nuclear density
rho = (5.0e17/1.78e-27)/100**3
m = array([1e0,1e40])
sigma_nuclear = 3*sqrt(pi)/4*(m/rho)**(2/3)
plt.plot(m,sigma_nuclear,'k:',alpha=0.3)
plt.text(1e21,0.7e-12,'Nuclear density',alpha=0.3,fontsize=20,rotation=19)



MySaveFig(fig,'Ultraheavy_Larger')
```
